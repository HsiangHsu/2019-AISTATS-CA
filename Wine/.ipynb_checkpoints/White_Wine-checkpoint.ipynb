```python
df_red = pd.read_csv('winequality-white-clean.csv')
df_red = df_red.iloc[:, 1:]
df_red.head()
```

```python
df_red.shape
```

```python
features = df_red.columns.values[:-1]
features
```

```python
X = df_red.iloc[:, :-1].values
labels = df_red.iloc[:, -1].values-1
y = pd.get_dummies(df_red.iloc[:, -1]).values
```

```python
cutoff = int(X.shape[0]*0.7)
X_train = X[:cutoff, :]
X_test = X[cutoff:, :]
y_train = y[:cutoff]
y_test = y[cutoff:]
train_labels = labels[:cutoff]
test_labels = labels[cutoff:]

X_train = (X_train-X_train.mean(0)) / X_train.std(0)
X_test = (X_test-X_test.mean(0)) / X_test.std(0)
```

```python
d = 6
dx = 11
dy = 7
session = tf.InteractiveSession()

(x_input,f_output,keepX) = PICnet(dx,name="Fnet",structure = [500, 100, 30, d])
(y_input,g_output,keepY) = PICnet(dy,name="Gnet",structure = [10, 5, 3, d])

objective,prodGiFG = create_loss_svd(f_output,g_output)
step = tf.train.AdamOptimizer(1e-3).minimize(objective)

# train
session.run(tf.global_variables_initializer())
n_steps = 1000
# obj = []
# train_loss = []
# test_loss = []
# index = []

for k in tqdm(range(n_steps)):
    session.run(step,feed_dict= {x_input:X_train,y_input:y_train})
#     if k%10==0:
#         index.append(k)
#         obj_train = objective.eval(feed_dict= {x_input:X_train,y_input:y_train})
#         train_loss.append(obj_train)
#         obj_test = objective.eval(feed_dict= {x_input:X_test,y_input:y_test})
#         .append(obj_test)
#         if np.abs(obj_test - obj_train) < 0.5:
#             break

    
    
    
# plt.plot(index, np.array(obj))

# plt.xlabel('Iteration')
# plt.ylabel('Loss')
```

```python
F_train_out = f_output.eval(feed_dict= {x_input:X_train,y_input:y_train})
G_train_out = g_output.eval(feed_dict= {x_input:X_train,y_input:y_train})
```

```python
A,a,B,b = normalizeFG(F_train_out, G_train_out)
F_train = (F_train_out-a).dot(A)
G_train = (G_train_out-b).dot(B)

pic_train = np.diag(F_train.transpose().dot(G_train) / F_train.shape[0])

F_test_out = f_output.eval(feed_dict= {x_input:X_test,y_input:y_test})
G_test_out = g_output.eval(feed_dict= {x_input:X_test,y_input:y_test})

F_test = (F_test_out-a).dot(A)
G_test = (G_test_out-b).dot(B)

pic_test = np.diag(F_test.transpose().dot(G_test) / F_test.shape[0])
```

```python
print(pic_train)
print(pic_test)
```

```python
resolution = 50

all_feature_grid = {}
X_mean = X_train.mean(0)


for i in range(11):
    max_ = max(X_train[:, i])
    min_ = min(X_train[:, i])
    X_feature = np.zeros((resolution, 11))
    X_feature[:, i] = np.linspace(min_, max_, resolution)
    F_train_feature_out = f_output.eval(feed_dict= {x_input:X_feature})
    F_feature = (F_train_feature_out-a).dot(A)
    all_feature_grid[features[i]] = F_feature
```

```python
fig = plt.figure(figsize=(10, 10))
# ax = plt.subplot(1, 1, 1)
plt.axvline(0, color='k', linestyle='--')
plt.axhline(0, color='k', linestyle='--')
pf1 = 2
pf2 = 3


ls = 20
plt.tick_params(axis='both', which='major', labelsize=ls)



for i in range(7):
    r, g, b = np.random.uniform(0, 1, 3)
    idx = np.arange(y_train.shape[0])[y_train[:, i] == 1]
    plt.plot(F_train[idx, pf1], F_train[idx, pf2], color=sns.xkcd_rgb[palette[i]], marker='.', linestyle='', alpha=0.8)
for i in range(7):
#     r, g, b = np.random.uniform(0, 1, 3)
    idx = np.arange(y_train.shape[0])[y_train[:, i] == 1]
    plt.plot(G_train[idx[0], pf1], G_train[idx[0], pf2], color=sns.xkcd_rgb[palette[i]], marker='s', linestyle='', label='Quality '+str(2.0+i))

cmaps = ['dark green', 'teal', 'dark orange', 'turquoise', 'magenta', 'yellow',
          'brick', 'dark grey', 'dark blue', 'dark purple', 'violet']
for i in range(11):
    feat = features[i]
    plt.plot(all_feature_grid[feat][:, pf1], all_feature_grid[feat][:, pf2], color=sns.xkcd_rgb[cmaps[i]], 
                  linestyle='--', markersize=10, label=feat, alpha = .5)
    plt.plot(all_feature_grid[feat][-1, pf1], all_feature_grid[feat][-1, pf2], color=sns.xkcd_rgb[cmaps[i]], 
                  linestyle='--', markersize=10, alpha = 1, marker='*')
#     plt.arrow(all_feature_grid[feat][-2, pf1], all_feature_grid[feat][-2, pf2], 
#               all_feature_grid[feat][-1, pf1]/30, 
#               all_feature_grid[feat][-1, pf2]/30,
#               color=sns.xkcd_rgb[cmaps[i]], linestyle='--', alpha = .7, 
#               width=0.05, shape='full')
#     for j in range(resolution-1):
#         patch = patches.FancyArrowPatch([all_feature_grid[feat][j, pf1], all_feature_grid[feat][j, pf2]], 
#                  [all_feature_grid[feat][j+1, pf1], all_feature_grid[feat][j+1, pf2]],
#                  color=sns.xkcd_rgb[cmaps[i]], 
#                  linestyle='-')
#         ax.add_patch(patch)

plt.tight_layout()
plt.legend(loc='best', ncol=2)
plt.xlabel('First Principal Function ({:.2f}%)'.format(pic_train[pf1]*100), fontsize=ls)
plt.ylabel('Second Principal Function ({:.2f}%)'.format(pic_train[pf2]*100), fontsize=ls)
# idx = np.arange(y_train.shape[0])[y_train[:, 0] == 1]
# plt.plot(G_train[idx[0], 0], G_train[idx[0], 1], color=sns.xkcd_rgb['blue'], marker='s', linestyle='')
# idx = np.arange(y_train.shape[0])[y_train[:, 1] == 1]
# plt.plot(G_train[idx[0], 0], G_train[idx[0], 1], color=sns.xkcd_rgb['dark green'], marker='s', linestyle='')
```

```python
def plot_wine(ax, pf1, pf2, F, G, y, all_feature_grid, features):
    ax.axvline(0, color='k', linestyle='--')
    ax.axhline(0, color='k', linestyle='--')
    ax.tick_params(axis='both', which='major', labelsize=ls)
    for i in range(6):
        idx = np.arange(y.shape[0])[y[:, i] == 1]
        ax.plot(F[idx, pf1], F[idx, pf2], color=sns.xkcd_rgb[palette[i]], marker='.', linestyle='', alpha=0.8)
    
    for i in range(6):
        idx = np.arange(y.shape[0])[y[:, i] == 1]
        ax.plot(G[idx[0], pf1], G[idx[0], pf2], color=sns.xkcd_rgb[palette[i]], marker='s', linestyle='', label='Quality '+str(2.0+i))

    for i in range(len(features)):
        feat = features[i]
        ax.plot(all_feature_grid[feat][:, pf1], all_feature_grid[feat][:, pf2], color=sns.xkcd_rgb[cmaps[i]], 
                      linestyle='--', markersize=10, label=feat, alpha = .5)
        ax.plot(all_feature_grid[feat][-1, pf1], all_feature_grid[feat][-1, pf2], color=sns.xkcd_rgb[cmaps[i]], 
                      linestyle='--', markersize=10, alpha = 1, marker='*')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ls = 20
cmaps = ['dark green', 'teal', 'dark orange', 'turquoise', 'magenta', 'yellow',
          'brick', 'dark grey', 'dark blue', 'dark purple', 'violet']

pf1, pf2 = 1, 2
plot_wine(ax[0], pf1, pf2, F_train, G_train, y_train, all_feature_grid, features)
ax[0].legend(loc='best', ncol=2, fontsize=13.5)
ax[0].set_xlabel('First Principal Function ({:.2f}%)'.format(pic_train[pf1]*100), fontsize=ls)
ax[0].set_ylabel('Second Principal Function ({:.2f}%)'.format(pic_train[pf2]*100), fontsize=ls)

pf1, pf2 = 2, 3
plot_wine(ax[1], pf1, pf2, F_train, G_train, y_train, all_feature_grid, features)
ax[1].set_xlabel('Second Principal Function ({:.2f}%)'.format(pic_train[pf1]*100), fontsize=ls)
ax[1].set_ylabel('Third Principal Function ({:.2f}%)'.format(pic_train[pf2]*100), fontsize=ls)

plt.tight_layout()
plt.savefig('ca_red_wine.pdf', format='pdf', dpi=300)
```

```python
''' PCA '''
from sklearn.decomposition import PCA
pca = PCA(n_components=6, whiten=True)
features_pca = pca.fit_transform(X_train, y_train)
print(pca.explained_variance_ratio_)
```

```python
''' KPCA '''
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=6, kernel='rbf')
features_kpca = kpca.fit_transform(X_train, y_train)
explained_variance = np.var(features_kpca, axis=0)
explained_variance_ratio = explained_variance / np.sum(explained_variance)
print(explained_variance_ratio)
```

```python
''' CCA '''
from sklearn.cross_decomposition import CCA
cca = CCA(n_components=6)
cca.fit(X_train, y_train)
X_c, Y_c = cca.transform(X_train, y_train)
corr_cca = X_c.transpose().dot(Y_c)/X_c.shape[0]
corr_cca = corr_cca/corr_cca.sum()
print(np.diag(corr_cca))
```

```python
''' KCCA '''
import rcca
rcca.degree = 4
kcca = rcca.CCA(kernelcca = True, reg = 5e-1, numCC = 6, ktype='poly')
kcca.train([X_train, y_train])
kX = kcca.comps[0]
kY = kcca.comps[1]
kcca_corr = np.diag(kX.transpose().dot(kY))
kcca_corr = kcca_corr / kcca_corr.sum()
print(kcca_corr)
```
