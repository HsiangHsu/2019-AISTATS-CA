```python
import numpy as np
import scipy as sp
import tensorflow as tf
import sys
import os
import pandas as pd
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
from fgnet import create_loss_svd, computeMetrics, normalizeFG
from fgnet import SimpleNet as PICnet
import matplotlib.patches as patches
sns.set()
sns.set_style('white')
%matplotlib inline
```

```python
df_red = pd.read_csv('winequality-red-clean.csv')
df_red = df_red.iloc[:, 1:]
df_red.head()
```

```python
df_red.shape
```

```python
features = df_red.columns.values[:-1]
features
```

```python
X = df_red.iloc[:, :-1].values
labels = df_red.iloc[:, -1].values-1
y = pd.get_dummies(df_red.iloc[:, -1]).values
```

```python
cutoff = int(X.shape[0]*0.7)
X_train = X[:cutoff, :]
X_test = X[cutoff:, :]
y_train = y[:cutoff]
y_test = y[cutoff:]
train_labels = labels[:cutoff]
test_labels = labels[cutoff:]

X_train = (X_train-X_train.mean(0)) / X_train.std(0)
X_test = (X_test-X_test.mean(0)) / X_test.std(0)
```

```python
d = 6
dx = 11
dy = 6
session = tf.InteractiveSession()

(x_input,f_output,keepX) = PICnet(dx,name="Fnet",structure = [500, 100, 30, d])
(y_input,g_output,keepY) = PICnet(dy,name="Gnet",structure = [10, 5, 3, d])

objective,prodGiFG = create_loss_svd(f_output,g_output)
step = tf.train.AdamOptimizer(1e-3).minimize(objective)

# train
session.run(tf.global_variables_initializer())
n_steps = 1000
# obj = []
# train_loss = []
# test_loss = []
# index = []

for k in tqdm(range(n_steps)):
    session.run(step,feed_dict= {x_input:X_train,y_input:y_train})
#     if k%10==0:
#         index.append(k)
#         obj_train = objective.eval(feed_dict= {x_input:X_train,y_input:y_train})
#         train_loss.append(obj_train)
#         obj_test = objective.eval(feed_dict= {x_input:X_test,y_input:y_test})
#         .append(obj_test)
#         if np.abs(obj_test - obj_train) < 0.5:
#             break

    
    
    
# plt.plot(index, np.array(obj))

# plt.xlabel('Iteration')
# plt.ylabel('Loss')
```

```python
F_train_out = f_output.eval(feed_dict= {x_input:X_train,y_input:y_train})
G_train_out = g_output.eval(feed_dict= {x_input:X_train,y_input:y_train})
```

```python
A,a,B,b = normalizeFG(F_train_out, G_train_out)
F_train = (F_train_out-a).dot(A)
G_train = (G_train_out-b).dot(B)

pic_train = np.diag(F_train.transpose().dot(G_train) / F_train.shape[0])

F_test_out = f_output.eval(feed_dict= {x_input:X_test,y_input:y_test})
G_test_out = g_output.eval(feed_dict= {x_input:X_test,y_input:y_test})

F_test = (F_test_out-a).dot(A)
G_test = (G_test_out-b).dot(B)

pic_test = np.diag(F_test.transpose().dot(G_test) / F_test.shape[0])
```

```python
print(pic_train)
print(pic_test)
```

```python
resolution = 50

all_feature_grid = {}
X_mean = X_train.mean(0)


for i in range(11):
    max_ = max(X_train[:, i])
    min_ = min(X_train[:, i])
    X_feature = np.zeros((resolution, 11))
    X_feature[:, i] = np.linspace(min_, max_, resolution)
    F_train_feature_out = f_output.eval(feed_dict= {x_input:X_feature})
    F_feature = (F_train_feature_out-a).dot(A)
    all_feature_grid[features[i]] = F_feature
```

```python
all_feature_max_min = {}
for i in range(11):
    max_ = max(X_train[:, i])
    min_ = min(X_train[:, i])
    X_feature = np.zeros((2, 11))
    X_feature[0, 0] = min_
    X_feature[1, 0] = max_
    F_train_feature_out = f_output.eval(feed_dict= {x_input:X_feature})
    F_feature = (F_train_feature_out-a).dot(A)
    all_feature_max_min[features[i]] = F_feature
```

```python
# all_feature_interpolate = {}
resolution = 50
lambda_ = np.linspace(0, 1, resolution)

all_feature_interpolate = {}
max_ = X_train[list(train_labels).index(7), :]
min_ = X_train[list(train_labels).index(2), :]

for i in range(resolution):
    X_feature = np.zeros((1, 11))
    X_feature[0, :] = lambda_[i]*max_ + (1-lambda_[i])*min_
    
    F_train_feature_out = f_output.eval(feed_dict= {x_input:X_feature})
    F_feature = (F_train_feature_out-a).dot(A)
    all_feature_interpolate[i] = F_feature
```

```python
def plot_wine_2(ax, pf1, pf2, F_train, G_train, all_feature_grid, features):
    ax.axvline(0, color='k', linestyle='--')
    ax.axhline(0, color='k', linestyle='--')
    ax.tick_params(axis='both', which='major', labelsize=ls)
    for i in range(6):
        idx = np.arange(y_train.shape[0])[y_train[:, i] == 1]
        ax.plot(F_train[idx, pf1], F_train[idx, pf2], color=sns.xkcd_rgb[palette[i]], marker='.', linestyle='', alpha=0.8)
    
    for i in range(6):
        idx = np.arange(y_train.shape[0])[y_train[:, i] == 1]
        ax.plot(G_train[idx[0], pf1], G_train[idx[0], pf2], color=sns.xkcd_rgb[palette[i]], marker='s', linestyle='', label='Quality '+str(2.0+i))
    for i in range(len(features)):
        feat = features[i]
        ax.plot(all_feature_grid[feat][0, pf1], all_feature_grid[feat][0, pf2], color=sns.xkcd_rgb[cmaps[i]], 
                marker='^', markersize=10, label='low '+feat, alpha = .9, linestyle='')
        ax.plot(all_feature_grid[feat][-1, pf1], all_feature_grid[feat][-1, pf2], color=sns.xkcd_rgb[cmaps[i]], 
                marker='>', markersize=10, label='high '+feat, alpha = .9, linestyle='')
    
    temp1 = []
    temp2 = []
    for i in range(len(all_feature_interpolate)):
        temp1.append(all_feature_interpolate[i][0, pf1])
        temp2.append(all_feature_interpolate[i][0, pf2])
    ax.plot(temp1, temp2, color=sns.xkcd_rgb['orange'], linestyle='--', label='Interpolation')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ls = 20
cmaps = ['dark green', 'teal', 'dark orange', 'turquoise', 'magenta', 'yellow',
          'brick', 'dark grey', 'dark blue', 'dark purple', 'violet']

pf1, pf2 = 1, 2
plot_wine_2(ax[0], pf1, pf2, F_train, G_train, all_feature_grid, features)
ax[0].legend(loc='best', ncol=3, fontsize=11)
ax[0].set_xlabel('First Principal Function ({:.2f}%)'.format(pic_train[pf1]/sum(pic_train)*100), fontsize=ls)
ax[0].set_ylabel('Second Principal Function ({:.2f}%)'.format(pic_train[pf2]/sum(pic_train)*100), fontsize=ls)

pf1, pf2 = 2, 3
plot_wine_2(ax[1], pf1, pf2, F_train, G_train, all_feature_grid, features)
ax[1].set_xlabel('Second Principal Function ({:.2f}%)'.format(pic_train[pf1]/sum(pic_train)*100), fontsize=ls)
ax[1].set_ylabel('Third Principal Function ({:.2f}%)'.format(pic_train[pf2]/sum(pic_train)*100), fontsize=ls)

plt.tight_layout()
```
